```python
# training the neural network on this simple data to understand the concept of RNN as it is a sequential data
# previous char i/p next char o/p - kinda thing in RNN 
alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
# enumerate - assigns the sequence no to perticular element in the string
list(enumerate(alphabets))
```

```python
# preparing input and output data using this list
'''2 datasets required - 1.encode string into numerical form 2.to decode the result into string againat the time of prediction - for this
purpuse dictionary comprehension method is used''';
```

```python
for i, c in enumerate(alphabets):
    print(i, c)
```

```python
# for encoding
int_to_char = dict((i,c) for i, c in enumerate(alphabets))
```

```python
int_to_char
```

```python
# for decoding
char_to_int = dict((c, i) for i, c in enumerate(alphabets))
```

```python
char_to_int
```

```python
#creating sequential i/p o/p - e.g. 1st char is i/p 2nd char will be o/p
# Z will not be i/p and A will not be o/p anywhere
seq_length = 1
x = []
y = []

for i in range (len(alphabets)-1):
    seq_in = alphabets[i:i+seq_length]
    seq_out = alphabets[i+seq_length]
    print(seq_in, '->' , seq_out)
    x.append([char_to_int[char] for char in seq_in]) # appended to the list as we want it in the list format
    y.append(char_to_int[seq_out])
```

```python
x
```

```python
y
```

```python
import numpy as np
```

```python
# necessary step in neural network
X = np.reshape(x, (len(x), seq_length, 1))
# 1 - how many number of characters are present in single sequence
# x- that we want to reshape
# len - sequence length
```

```python
X.shape
```

```python
X
```

```python
# scaling - shifting the data in the range of 0 to 1 without changing the internal relation
a = np.array([7,3,0,2,4,8,5])
```

```python
# dividing by highest no in the list
# this step is to speed up th process of ML
# here only magnitudes are changed
a / 8
```

```python
#preperation of input data
```

```python
X = X / 25
```

```python
X
```

```python
# preperation of output data
# one-hot encoding- 
```

```python
a = np.array([3,0,2,3,3,1,2,3,1,0,2,0,0,0,1,2,3])
```

```python
a
```

```python
from keras.utils import to_categorical
```

```python
#requirement of multiclass classification
to_categorical(a)
```

```python
# one-hot encoding for o/p data
y = to_categorical(y)
```

```python
## build the model
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
#Create the object of RNN
model = Sequential()

#add the LSTM layer
layer1 = LSTM(32, input_shape = (X.shape[1], X.shape[2])) #first parameter - how many bytes are required i.e. memory, 2nd para - i/p shape - 1 x 1
model.add(layer1)

#add the o/p layer
layer2 = Dense(26, activation='softmax')
model.add(layer2)
```

```python
model.summary()
```

```python
from keras.utils import plot_model
```

```python
plot_model(model, show_layer_activations=True, show_layer_names=True, show_shapes=True)
```

```python
#Compile the model
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
## train the model
```

```python
model.fit(X, y, epochs=1000, batch_size=1)
```

```python
model.evaluate(X, y, batch_size=1)
```
